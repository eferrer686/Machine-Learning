```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
bikes = pd.read_csv("insurance.csv")
bikes
```

```python
from numpy import genfromtxt
bikes = genfromtxt('insurance.csv', delimiter=',')

# np.random.shuffle(bikes)

X = bikes[1:,0:6]
Y = bikes[1:,[6]]

print(X)
print(Y)
```

```python
for cont in range(6):
    X[:,cont] = ((X[:,cont]-X[:,cont].min())/(X[:,cont].max()-X[:,cont].min()))
    
for cont in range(1):
    Y[:,cont] = (Y[:,cont]/(Y[:,cont].max()))
    
print(X)
print(Y)
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(6, activation=tf.nn.relu),
  tf.keras.layers.Dense(7, activation=tf.nn.relu),
  tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

```

```python
model.fit(X, Y, epochs=15)
```

```python
scores = model.evaluate(X, Y)
print(scores)
```

```python
scores = model.predict(X)
print(scores)
```
